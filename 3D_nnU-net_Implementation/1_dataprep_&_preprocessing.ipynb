```python
from google.colab import drive
import os


drive.mount('/content/drive')


input_images_dir = "/content/drive/MyDrive/BreastDM_output/images"
input_masks_dir = "/content/drive/MyDrive/BreastDM_output/masks"

output_dir = "/content/drive/MyDrive/BreastDM_nnUNet/Dataset001_BreastCancer"
output_imagesTr = os.path.join(output_dir, "imagesTr")
output_labelsTr = os.path.join(output_dir, "labelsTr")


os.makedirs(output_imagesTr, exist_ok=True)
os.makedirs(output_labelsTr, exist_ok=True)

print(f"✅ Setup complete! Output will be saved in Google Drive: {output_dir}")
```

```python
import os
import numpy as np
import nibabel as nib

#CONVERTING FILES FROM .npy TO .nii.gz
patient_files = sorted([f for f in os.listdir(input_images_dir) if f.endswith(".npy")])


for idx, patient_file in enumerate(patient_files):
    patient_id = f"BC_{idx:03d}"


    image_data = np.load(os.path.join(input_images_dir, patient_file))
    mask_data = np.load(os.path.join(input_masks_dir, patient_file))

    slices = image_data.shape[0]

    #for image, taking all 3 modalities
    for mod_idx, mod_name in enumerate(["SUB2", "VIBRANT", "VIBRANT+C2"]): #mod_idx iterates over each mod_name and we each modality stored in nii_image
        nii_image = nib.Nifti1Image(image_data[:, mod_idx], affine=np.eye(4))
        nib.save(nii_image, os.path.join(output_imagesTr, f"{patient_id}_{mod_idx:04d}.nii.gz")) #converting to nii.gz format

    #for masks, taking only SUB2.npy modality
    nii_mask = nib.Nifti1Image(mask_data[:, 0], affine=np.eye(4)) #here, we have just [:,0] as we are just taking the SUB2.npy
    nib.save(nii_mask, os.path.join(output_labelsTr, f"{patient_id}.nii.gz"))

    print(f"✅ Converted {patient_file} → {slices} slices saved as {patient_id}.")

print("🎉 All files converted successfully!")
```

```python
import os
import nibabel as nib
import numpy as np


labels_dir = "/content/drive/MyDrive/BreastDM_nnUNet/Dataset001_BreastCancer/labelsTr"


for file in os.listdir(labels_dir):
    if file.endswith(".nii.gz"):
        file_path = os.path.join(labels_dir, file)


        nii = nib.load(file_path)
        data = nii.get_fdata()

        #binarizing the labels data to have only values 0 and 1.
        data[data == 255] = 1


        new_nii = nib.Nifti1Image(data.astype(np.int64), nii.affine, nii.header)
        nib.save(new_nii, file_path)
        print(f"Fixed {file_path}")

print("All label files updated successfully.")
```
